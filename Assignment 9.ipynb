```python
#2) Create a DatetimeIndex that contains each business day of 2015 and use it to index a Series of random numbers.

#Import Pandas and Numpy
import pandas as pd
import numpy as np
# Load all the business days of 2015 into duration
duration = pd.date_range('2015-01-01', '2015-12-31', freq='B')
# make it as date time index
idx = pd.DatetimeIndex(duration)
# generate a series of random numbers for each business days
row_labels = np.random.randint(1,261,size=(261,))

# create a data frame with business dates and random numbers
df = pd.DataFrame({'my_dates':idx,'myvals':row_labels})
# Insert day of week each days like Monday, Tuesday, etc.
df['day_of_week'] = df['my_dates'].dt.weekday_name
# Insert month each days like 1, 2, 3, etc.
df['month'] = df['my_dates'].dt.month
#df_dateindex = pd.DatetimeIndex(df)
df.set_index('my_dates',inplace=True)
df
```

```python
# 3) Find the sum of the values in s for every Wednesday

# Load all wednesday data into a
a = df.loc[df['day_of_week'] == 'Wednesday']
# Get sum of random numbers of all wednesdays in 2015
count_wed = a['myvals'].sum()
#Print count
count_wed
```

```python
# 4) Average For each calendar month

average = df.groupby('month').myvals.agg(['mean'])
average
```

```python
# 5) For each group of four consecutive calendar months in s, find the date on which the highest value occurred.
# separate three groups of four consecutive calender months
g1 = df[(df.month == 1) | (df.month == 2)| (df.month == 3)| (df.month == 4)]
g2 = df[(df.month == 5) | (df.month == 6)| (df.month == 7)| (df.month == 8)]
g3 = df[(df.month == 9) | (df.month == 10)| (df.month == 11)| (df.month == 12)]
# Find maximum of three groups
g1_m = g1['myvals'].max()
g2_m = g2['myvals'].max()
g3_m = g3['myvals'].max()
# print the date of maximum value
print(g1.index[g1['myvals'] == g1_m].tolist())
print(g2.index[g2['myvals'] == g2_m].tolist())
print(g3.index[g3['myvals'] == g3_m].tolist())
#df.get_loc(g1_m, method=None, tolerance=None)
```
